```python
def sumasentimientos(data_sentimientos, data_tweets):
    try:
    #Archivo Sentimientos
        sentimientos = open(data_sentimientos)
        valores = {} 
        for linea in sentimientos: 
            termino, valor = linea.split("\t") 
            valores[termino] = int(valor)
        
    
    #Archivo Tweets
        import json
        with open (data_tweets, 'r') as archivotweets:
            datos_noficial=archivotweets.readlines()
            s=[]
            for linea in datos_noficial:
                try:
                    s1=json.loads(linea)
                    s.append(s1)
                except json.JSONDecodeError as e:
                    print(f"Error al decodificar JSON en la lÃ­nea: {linea}")
                    print(e)
#print(s)
        base_tw={}
        i=0
        for datos in s:
            base_tw[i]=datos.get('text','N/A').lower()
            i+=1
#   bse_tw=datos.get('text','N/A')
    #####Programacion de suma de sentimientos
        diccionario_rf={}
        for tw_values in base_tw.values():
            puntos=0
            if tw_values!='n/a':
                texto=tw_values.split(" ")
                for valida in texto:
                    valida2=valida.strip('#')
                    puntos+=valores.get(valida2,0)
            diccionario_rf[tw_values]=puntos
#            diccionario_rf.pop('n/a')
        #print(f"{diccionario_rf.keys()} TIENE UN SENTIMIENTO ASOSIADO {diccionario_rf.values()}")
#        print(diccionario_rf)
#No estoy segura por el filtrado de tweets sin valor asociado.        
        for clave, valores in diccionario_rf.items():
            if clave!='n/a':
                print(f"{clave} TIENE UN SENTIMIENTO ASOSIADO {valores}")
#            elif clave!='n/a' and valores == 0: 
#                print(f"{clave} TIENE UN SENTIMIENTO ASOSIADO 0")
        archivotweets.close()
        sentimientos.close()
    
    except FileNotFoundError :
        print("Error: Los archivos no se encuentran")
    except Exception as e:
        print(f"Error:{e}")
    

```

```python
data_sentimientos= input("Introduzca el path del fichero que contiene el valor de los sentimientos para los terminos: ")
data_tweets= input("Introduzca el path del fichero que contiene los tweets:")
sumasentimientos(data_sentimientos, data_tweets)
```
